## 0.1 PySpark data loading and preliminary cleaning
This module handles the initial ingestion and cleaning of the raw dataset using PySpark. It loads the UCI Household Power Consumption data, replaces missing value placeholders (?) with nulls, removes invalid rows, and casts all measurement fields to DoubleType. A proper Datetime column is created by combining the date and time strings, enabling accurate time-based operations.

After cleaning, the resulting DataFrame is stored in Parquet format for efficient downstream processing.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, concat_ws, to_timestamp
from pyspark.sql.types import DoubleType
```

```python
# Initialize Spark session
spark = SparkSession.builder \
    .appName("Electric Power Project - Data Cleaning") \
    .getOrCreate()

# File path
file_path = "/content/drive/MyDrive/electric_power_project/data/raw/household_power.csv"

# Load data (semicolon-separated, '?' as missing)
df_raw = spark.read.csv(file_path, header=True, sep=";", inferSchema=False)

# Replace '?' with null and drop rows with nulls
df = df_raw.replace("?", None).dropna()

# Cast numeric columns to DoubleType
columns_to_cast = [
    "Global_active_power", "Global_reactive_power", "Voltage",
    "Global_intensity", "Sub_metering_1", "Sub_metering_2", "Sub_metering_3"
]
for col_name in columns_to_cast:
    df = df.withColumn(col_name, col(col_name).cast(DoubleType()))

# Create timestamp column from Date and Time
df = df.withColumn("Datetime", to_timestamp(concat_ws(" ", col("Date"), col("Time")), "d/M/yyyy H:mm:ss"))

# Drop original Date and Time columns
df = df.drop("Date", "Time")

# Reorder columns (Datetime first)
cols = df.columns
cols.remove("Datetime")
df = df.select(["Datetime"] + cols)

# Show schema and sample data
df.printSchema()
df.show(5, truncate=False)

# Save cleaned data to Parquet
output_path = "/content/drive/MyDrive/electric_power_project/data/processed/cleaned_power.parquet"
df.write.mode("overwrite").parquet(output_path)

print("Data cleaning completed and saved to:", output_path)
```

## 0.2 Aggregate data to daily level using PySpark
Convert the cleaned minute-level electricity data into daily-level aggregated data to be used in time series forecasting and modeling. Key metrics such as average and total active power, voltage statistics, intensity, and sub-metering sums are computed for each day.

This step reduces data granularity while preserving meaningful consumption patterns, and forms the primary dataset for supervised modeling.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import to_date, avg, max, min, sum as spark_sum

# Initialize Spark session
spark = SparkSession.builder \
    .appName("Electric Power Project - Daily Aggregation") \
    .getOrCreate()

# Load the cleaned Parquet data
input_path = "/content/drive/MyDrive/electric_power_project/data/processed/cleaned_power.parquet"
df = spark.read.parquet(input_path)

# Extract date from timestamp
df = df.withColumn("Date", to_date("Datetime"))

# Aggregate data by date using common statistical functions
df_daily = df.groupBy("Date").agg(
    avg("Global_active_power").alias("avg_active_power"),
    spark_sum("Global_active_power").alias("total_active_power"),
    avg("Voltage").alias("avg_voltage"),
    max("Voltage").alias("max_voltage"),
    min("Voltage").alias("min_voltage"),
    avg("Global_intensity").alias("avg_intensity"),
    spark_sum("Sub_metering_1").alias("sum_sub_1"),
    spark_sum("Sub_metering_2").alias("sum_sub_2"),
    spark_sum("Sub_metering_3").alias("sum_sub_3")
)

# Optional: sort by date
df_daily = df_daily.orderBy("Date")

# Show results
df_daily.show(5)

# Save to Parquet for modeling
output_path = "/content/drive/MyDrive/electric_power_project/data/processed/daily_power.parquet"
df_daily.write.mode("overwrite").parquet(output_path)

print("Daily-level data saved to:", output_path)
```

## 0.4 Feature Engineering for Regression
To improve prediction accuracy, a series of time-based features were engineered from daily electricity consumption data:

**Target Variable**: Defined as the 3-day rolling average of next day's total consumption to smooth out short-term fluctuations.

**Lag Features**: Power values from the previous 1–7 days to capture short-term dependencies.

**Rolling Statistics**: 7-day rolling mean and standard deviation (excluding the current day) to reflect local trends and variability.

**Temporal Indicators**: Day-of-week and weekend flags to model weekly patterns.

**Trend and Difference**: Cumulative sum and first-order difference to capture long-term drift and abrupt changes.

**Seasonal Encodings**: Sine and cosine transformations of the day of year to encode yearly seasonality in a cyclic manner.

Missing values from shifting/rolling operations were dropped, and the finalized dataset was saved for model training.

```python
import pandas as pd
import numpy as np
import os
```

```python
# Load daily-level data
file_path = "/content/drive/MyDrive/electric_power_project/data/processed/daily_power.parquet"
df = pd.read_parquet(file_path)

# Ensure datetime format
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)

# Target: predict next day's total consumption
df['target'] = df['total_active_power'].shift(-1).rolling(3).mean()

# Lag features
for lag in range(1, 8):
    df[f'lag_{lag}'] = df['total_active_power'].shift(lag)

# Rolling features
df['rolling_mean_7'] = df['total_active_power'].shift(1).rolling(window=7).mean()
df['rolling_std_7'] = df['total_active_power'].shift(1).rolling(window=7).std()

# Weekday features
df['day_of_week'] = df['Date'].dt.dayofweek
df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)

# Trend characteristics
df['cumsum_power'] = df['total_active_power'].cumsum()
df['diff_power'] = df['total_active_power'].diff()

# Cycle characteristics
df['month'] = df['Date'].dt.month
df['day_of_year'] = df['Date'].dt.dayofyear
df['sin_day'] = np.sin(2 * np.pi * df['day_of_year'] / 365)
df['cos_day'] = np.cos(2 * np.pi * df['day_of_year'] / 365)

# Drop rows with NaNs
df = df.dropna().reset_index(drop=True)

# Drop Date and intermediate columns
df_final = df.drop(columns=["Date", "day_of_year"])

# Save to CSV
output_path = "/content/drive/MyDrive/electric_power_project/data/processed/feature_data_v2.csv"
os.makedirs(os.path.dirname(output_path), exist_ok=True)
df_final.to_csv(output_path, index=False)

print("New feature data saved to:", output_path)
```

## 1.1 & 1.2 Random Forest & Gradient Boosted Trees (GBT)
### Model Training and Evaluation with PySpark Tree-Based Regressors

To assess the performance of traditional ensemble models on the electricity forecasting task, two tree-based regressors were implemented and evaluated using PySpark:

- **Dataset Handling:**
  - The enhanced dataset was loaded and transformed into a feature vector using `VectorAssembler`.
  - A random 80/20 train-test split was applied.

- **Models Trained:**

  - Both models were trained on the training data and saved to disk.

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import RandomForestRegressor, GBTRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml import Pipeline
from pyspark.ml.regression import RandomForestRegressionModel, GBTRegressionModel

import os
import pandas as pd
```

```python
# Create Spark session
spark = SparkSession.builder \
    .appName("ElectricPower Forecasting v3 - PySpark") \
    .getOrCreate()

# Load the enhanced dataset
df = spark.read.csv(
    "/content/drive/MyDrive/electric_power_project/data/processed/feature_data_v2.csv",
    header=True,
    inferSchema=True
)

# Assemble features
feature_cols = [col for col in df.columns if col != 'target']
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
df = assembler.transform(df).select("features", "target")

# Train/test split
train_data, test_data = df.randomSplit([0.8, 0.2], seed=42)

# Random Forest Regressor (tuned)
rf = RandomForestRegressor(
    labelCol="target", featuresCol="features",
    numTrees=200, maxDepth=10, minInstancesPerNode=4,
    featureSubsetStrategy="auto", seed=42
)

# GBT Regressor (tuned)
gbt = GBTRegressor(
    labelCol="target", featuresCol="features",
    maxIter=150, maxDepth=6, stepSize=0.1,
    minInstancesPerNode=4, seed=42
)

# Train models
rf_model = rf.fit(train_data)
gbt_model = gbt.fit(train_data)

# Save models
rf_model.save("/content/drive/MyDrive/electric_power_project/models/rf_model_v2")
gbt_model.save("/content/drive/MyDrive/electric_power_project/models/gbt_model_v2")

# Predict
rf_pred = rf_model.transform(test_data)
gbt_pred = gbt_model.transform(test_data)

# Save predictions
rf_pred.select("prediction", "target").toPandas().to_csv(
    "/content/drive/MyDrive/electric_power_project/results/rf_prediction_v2.csv", index=False
)
gbt_pred.select("prediction", "target").toPandas().to_csv(
    "/content/drive/MyDrive/electric_power_project/results/gbt_prediction_v2.csv", index=False
)

# Evaluate
evaluator = RegressionEvaluator(labelCol="target", predictionCol="prediction")
```

## 1.3 MLP Regression (Keras)
To explore the performance of deep learning on electricity consumption forecasting, a fully connected neural network (Multilayer Perceptron) was implemented using TensorFlow/Keras.

- **Data Preprocessing:**
  - The enhanced feature dataset was normalized using MinMaxScaler to ensure stable convergence during training.
  - The data was split into training and test sets with an 80/20 ratio.

- **Model Architecture:**
  - Input Layer: Dense layer with 128 units and ReLU activation.
  - Hidden Layer: Dense layer with 64 units, also using ReLU.
  - Each dense layer is followed by Batch Normalization and Dropout for regularization.
  - Output Layer: A single unit for regression output.

- **Training Setup:**
  - Optimizer: Adam with a learning rate of 0.001.
  - Loss: Mean Squared Error (MSE), with MAE tracked as an additional metric.
  - Early stopping was applied to prevent overfitting, monitoring validation loss with patience of 15 epochs.

- **Results & Export:**
  - The model was evaluated using standard regression metrics (RMSE, MAE, R²).
  - Predictions were saved for analysis, and the trained model was exported in `.keras` format for reproducibility.

This MLP provides a deep learning baseline to compare against tree-based models and LSTM architectures.

```python
import pandas as pd
import numpy as np
import os
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load feature-enhanced dataset
df = pd.read_csv("/content/drive/MyDrive/electric_power_project/data/processed/feature_data_v2.csv")

# Split features and target
X = df.drop(columns=["target"]).values
y = df["target"].values

# Normalize features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Define model structure
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.2),

    tf.keras.layers.Dense(1)
])

# Compile model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae']
)

# Add early stopping callback
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=15, restore_best_weights=True
)

# Train the model
history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=300,
    batch_size=16,
    callbacks=[early_stop],
    verbose=0  # Set to 1 to show training progress
)

# Predict
y_pred = model.predict(X_test).flatten()

# Evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Save predictions
df_result = pd.DataFrame({"target": y_test, "prediction": y_pred})
os.makedirs("/content/drive/MyDrive/electric_power_project/results", exist_ok=True)
df_result.to_csv("/content/drive/MyDrive/electric_power_project/results/mlp_keras_prediction_v2.csv", index=False)

# Save model
model.save("/content/drive/MyDrive/electric_power_project/models/mlp_keras_model_v2.keras")
```

## 1.4 LSTM (Time2Vec + BiLSTM, Multi-Output Averaging)

  - It transforms time information into a mix of linear and sinusoidal components, improving the temporal awareness of the model.

- **Model Architecture:**
  - Three stacked Bidirectional LSTM layers progressively extract hierarchical temporal patterns from the 60-day historical sequences.
  - Residual connections with dense projection layers were used to stabilize deep sequence learning.
  - Layer normalization and dropout were applied for improved convergence and regularization.
  - Dense layers at the output stage refine the learned temporal features for final prediction.

- **Training Setup:**
  - Input: Sliding windows of the previous 60 days (with engineered features), with the target being the **average power consumption over the next 3 days**.
  - Optimization: Adam optimizer with a learning rate schedule.
  - Early stopping and learning rate reduction callbacks were used to enhance generalization and avoid overfitting.

- **Results:**
  - The model achieved strong predictive performance based on RMSE, MAE, and R².
  - Results were saved for reporting, and the trained model was exported in Keras format.

This model serves as a key benchmark for long-range temporal forecasting and outperformed static models in capturing consumption dynamics.

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Time2Vec Layer
class Time2Vec(tf.keras.layers.Layer):
    def __init__(self, kernel_size=4):
        super(Time2Vec, self).__init__()
        self.kernel_size = kernel_size

    def build(self, input_shape):
        self.feature_dim = input_shape[-1]
        self.wb = self.add_weight(name='wb', shape=(self.feature_dim,), initializer='uniform', trainable=True)
        self.bb = self.add_weight(name='bb', shape=(self.feature_dim,), initializer='uniform', trainable=True)
        self.wa = self.add_weight(name='wa', shape=(self.kernel_size, self.feature_dim), initializer='uniform', trainable=True)
        self.ba = self.add_weight(name='ba', shape=(self.kernel_size, self.feature_dim), initializer='uniform', trainable=True)

    def call(self, inputs):
        bias = self.wb * inputs + self.bb
        expand = tf.expand_dims(inputs, axis=2)
        wa = tf.expand_dims(tf.expand_dims(self.wa, axis=0), axis=0)
        ba = tf.expand_dims(tf.expand_dims(self.ba, axis=0), axis=0)
        dp = tf.math.sin(expand * wa + ba)
        dp = tf.reshape(dp, shape=(-1, tf.shape(inputs)[1], self.kernel_size * self.feature_dim))
        return tf.concat([bias, dp], axis=-1)

# Model Definition
def build_lstm_single_output(n_steps, n_features):
    inputs = tf.keras.Input(shape=(n_steps, n_features))
    x = Time2Vec(kernel_size=4)(inputs)

    x1 = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences=True))(x)
    x1 = tf.keras.layers.LayerNormalization()(x1)
    x1 = tf.keras.layers.Dropout(0.3)(x1)

    x2 = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True))(x1)
    x2 = tf.keras.layers.LayerNormalization()(x2)
    x2 = tf.keras.layers.Dropout(0.2)(x2)

    if x1.shape[-1] != x2.shape[-1]:
        x1 = tf.keras.layers.Dense(x2.shape[-1])(x1)
    x = tf.keras.layers.Add()([x1, x2])

    x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32, return_sequences=False))(x)
    x = tf.keras.layers.LayerNormalization()(x)
    x = tf.keras.layers.Dropout(0.2)(x)

    x = tf.keras.layers.Dense(64, activation='relu')(x)
    x = tf.keras.layers.Dense(32, activation='relu')(x)
    outputs = tf.keras.layers.Dense(1)(x)

    model = tf.keras.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
                  loss='mse', metrics=['mae'])
    return model

# Data
df = pd.read_csv("/kaggle/input/electricity/feature_data_v2.csv")
if "Date" in df.columns:
    df = df.drop(columns=["Date"])

X_all = df.drop(columns=["target"])
y_all = df["target"].values

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_all)

n_steps = 60
pred_len = 3

def create_sequences_avg(X, y, step, pred_len=3):
    Xs, ys = [], []
    for i in range(len(X) - step - pred_len + 1):
        Xs.append(X[i:i+step])
        ys.append(np.mean(y[i+step:i+step+pred_len]))
    return np.array(Xs), np.array(ys)

X_seq, y_seq = create_sequences_avg(X_scaled, y_all, step=n_steps, pred_len=pred_len)

X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)

# Train
model = build_lstm_single_output(n_steps=n_steps, n_features=X_seq.shape[2])
model.summary()

callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, min_lr=1e-6)
]

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=300,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

# Evaluation
y_pred = model.predict(X_test).flatten()
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("LSTM Results:")
print("RMSE:", round(rmse, 3))
print("MAE :", round(mae, 3))
print("R²  :", round(r2, 3))

# Save
df_result = pd.DataFrame({"target": y_test, "prediction": y_pred})
df_result.to_csv("/kaggle/working/lstm_prediction_avg3_single.csv", index=False)
model.save("/kaggle/working/lstm_model_avg3_single.keras")
```

# 2. Model Refinement: Time-Series Feature Engineering & Anomaly Detection



## 2.1Time-Series Feature Engineering and Anomaly Detection

To further enhance model accuracy and robustness, advanced feature engineering and anomaly detection techniques were applied to the time-series dataset.

#### Rolling Statistics and Temporal Dynamics

Additional features were engineered based on the `target` variable (i.e., next-3-day average consumption):

- **Rolling Mean & Std (3-day window):**  
  - `mean_last_3_days`: Captures short-term average trends.
  - `std_last_3_days`: Quantifies local volatility in demand.

- **Delta Feature:**
  - `delta_from_yesterday`: Measures daily change compared to the previous day (`lag_1`), highlighting sudden consumption shifts.

These engineered features improve the model’s ability to learn **temporal dependencies and short-term irregularities**.

#### Hybrid Anomaly Detection

Two complementary anomaly detection methods were introduced to improve model generalization:

1. **Z-score Method:**
   - A statistical thresholding approach using the standardized deviation from the mean.
   - Values with absolute Z-score > 3 are marked as anomalies (`anomaly_zscore = 1`).

2. **Isolation Forest (Unsupervised Machine Learning):**
   - A tree-based ensemble method that isolates anomalies based on how easily they can be separated in feature space.
   - A binary label `anomaly_iforest` is generated, with `1` indicating anomalous patterns in multivariate space.

Both binary anomaly flags can be:
- Used as input features to improve **robustness against outliers**, or
- Employed to **exclude abnormal samples** from training during model tuning.

#### Output

The enriched dataset (`feature_data_v3.csv`) now incorporates:
- Temporal patterns
- Local trends and variability
- Statistical and model-based anomaly indicators

This version serves as the final feature base for models targeting high-accuracy predictions under real-world noisy data conditions.


```python
import pandas as pd

#Load original data
df = pd.read_csv("/content/drive/MyDrive/electric_power_project/data/processed/feature_data_v2.csv")

# Time-Series Feature Engineering
# Add 3-day rolling mean and standard deviation of target
df['mean_last_3_days'] = df['target'].rolling(window=3).mean()
df['std_last_3_days'] = df['target'].rolling(window=3).std()

# Add delta from previous day
df['delta_from_yesterday'] = df['target'] - df['lag_1']

# Drop rows with NaN after rolling window
df = df.dropna().reset_index(drop=True)
```

```python
# Anomaly Detection using Z-score
z_thresh = 3
df['zscore'] = (df['target'] - df['target'].mean()) / df['target'].std()
df['anomaly_zscore'] = (df['zscore'].abs() > z_thresh).astype(int)

# Anomaly Detection using Isolation Forest
from sklearn.ensemble import IsolationForest

# Select features for Isolation Forest
features_for_iforest = [
    'target', 'lag_1', 'lag_2', 'rolling_mean_7', 'rolling_std_7',
    'avg_active_power', 'avg_voltage', 'avg_intensity'
]

# Fit Isolation Forest model
iso = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
df['anomaly_iforest'] = iso.fit_predict(df[features_for_iforest])

# Convert to binary anomaly flag
df['anomaly_iforest'] = (df['anomaly_iforest'] == -1).astype(int)

output_path = "/content/drive/MyDrive/electric_power_project/data/processed/feature_data_v3.csv"
df.to_csv(output_path, index=False)

output_path
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Set result directory path
result_dir = "/content/drive/MyDrive/electric_power_project/results"
files = {
    "MLP": "mlp_keras_prediction_v3.csv",
    "Random Forest": "rf_prediction_v3.csv",
    "GBT": "gbt_prediction_v3.csv",
    "XGBoost": "xgb_prediction_v3.csv"
}

# Store evaluation results
results = []

for name, file in files.items():
    path = os.path.join(result_dir, file)
    df = pd.read_csv(path)

    y_true = df["target"]
    y_pred = df["prediction"]

    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    nrmse = rmse / (y_true.max() - y_true.min())

    results.append({
        "Model": name,
        "RMSE": round(rmse, 3),
        "MAE": round(mae, 3),
        "R²": round(r2, 4),
        "NRMSE": round(nrmse, 4)
    })

# Convert to DataFrame
df_metrics = pd.DataFrame(results)

# Print evaluation table
print(df_metrics)
```

### Optimal Cluster Selection via Silhouette Score

To determine the optimal number of clusters (`k`), we computed the **Silhouette Score** for values ranging from 2 to 9. The silhouette score evaluates the cohesion and separation of clusters, with higher values indicating better-defined clusters.

As shown in the figure, the silhouette score peaks at **k = 2** with a value of approximately **0.206**. This suggests that the dataset naturally separates into two distinct groups. Increasing the number of clusters beyond 2 leads to a sharp decline in silhouette scores, implying reduced clustering quality.

### Dimensionality Reduction and Cluster Visualization

To visualize the high-dimensional clustering results, we applied two nonlinear dimensionality reduction techniques:

- **t-SNE (t-distributed Stochastic Neighbor Embedding)**
- **UMAP (Uniform Manifold Approximation and Projection)**

We projected the clusters obtained from both **KMeans** and **Agglomerative Clustering** onto 2D space using these techniques.

- In **t-SNE plots**, both clustering algorithms form distinguishable but partially overlapping clusters.
- In **UMAP plots**, the clusters appear more compact and well-separated, especially for KMeans, suggesting UMAP is better at preserving local and global structure in this case.

### Summary of Findings

- **k = 2** is the optimal number of clusters based on silhouette analysis.
- Both KMeans and Agglomerative clustering effectively separate the data into meaningful groups.
- **UMAP visualization** shows clearer cluster boundaries compared to t-SNE.
- The identified clusters could represent different electricity usage patterns (e.g., seasonal or weekday vs. weekend).
- These clusters can be used for **cluster-aware forecasting**, enabling more targeted and accurate modeling strategies per group.

```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE
import umap.umap_ as umap
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load and prepare the dataset (v3)
df = pd.read_csv("/content/drive/MyDrive/electric_power_project/data/processed/feature_data_v3.csv")
X = df.drop(columns=["target"])

# 2. Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Find best k for KMeans using silhouette score
k_range = range(2, 10)
silhouettes = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    labels = km.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouettes.append(score)
best_k = k_range[np.argmax(silhouettes)]

# Save best_k to log file
log_path = "/content/drive/MyDrive/electric_power_project/results/clustering_log.txt"
with open(log_path, "w") as f:
    f.write(f"Best k by silhouette score: {best_k}\n")
    f.write("Silhouette scores by k:\n")
    for k, score in zip(k_range, silhouettes):
        f.write(f"k={k}: {score:.4f}\n")

# Plot silhouette score vs k
plt.figure(figsize=(8, 5))
plt.plot(list(k_range), silhouettes, marker='o', linestyle='-', color='blue')
plt.axvline(x=best_k, color='red', linestyle='--', label=f"Best k = {best_k}")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs. Number of Clusters")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("/content/drive/MyDrive/electric_power_project/results/silhouette_vs_k_v3.png")
plt.show()

# 4. Fit KMeans and Agglomerative clustering with best_k
kmeans = KMeans(n_clusters=best_k, random_state=42)
df["kmeans_cluster"] = kmeans.fit_predict(X_scaled)

agg = AgglomerativeClustering(n_clusters=best_k)
df["hier_cluster"] = agg.fit_predict(X_scaled)
```

```python
# 5. Dimensionality reduction: t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)
df["tsne_1"] = X_tsne[:, 0]
df["tsne_2"] = X_tsne[:, 1]

# 6. Dimensionality reduction: UMAP
umap_reducer = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_reducer.fit_transform(X_scaled)
df["umap_1"] = X_umap[:, 0]
df["umap_2"] = X_umap[:, 1]

# 7. Visualization of clustering results
sns.set(style="whitegrid", font_scale=1.1)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# KMeans - t-SNE
sns.scatterplot(x="tsne_1", y="tsne_2", hue="kmeans_cluster", palette="tab10", data=df, ax=axs[0, 0], s=30)
axs[0, 0].set_title("t-SNE - KMeans Clustering")

# Hierarchical - t-SNE
sns.scatterplot(x="tsne_1", y="tsne_2", hue="hier_cluster", palette="tab10", data=df, ax=axs[0, 1], s=30)
axs[0, 1].set_title("t-SNE - Hierarchical Clustering")

# KMeans - UMAP
sns.scatterplot(x="umap_1", y="umap_2", hue="kmeans_cluster", palette="tab10", data=df, ax=axs[1, 0], s=30)
axs[1, 0].set_title("UMAP - KMeans Clustering")

# Hierarchical - UMAP
sns.scatterplot(x="umap_1", y="umap_2", hue="hier_cluster", palette="tab10", data=df, ax=axs[1, 1], s=30)
axs[1, 1].set_title("UMAP - Hierarchical Clustering")

plt.tight_layout()
plt.savefig("/content/drive/MyDrive/electric_power_project/results/clustering_tsne_umap_v3.png")
plt.show()

# 8. Save clustered dataset (excluding t-SNE/UMAP)
df.drop(columns=["tsne_1", "tsne_2", "umap_1", "umap_2"]).to_csv(
    "/content/drive/MyDrive/electric_power_project/data/processed/feature_data_clustered_v3.csv",
    index=False
)
```

## Identify distinct electricity consumption patterns

## Cluster-wise Electricity Consumption Profiles (Radar Chart Analysis)

To further interpret the behavioral patterns of electricity consumption revealed by KMeans clustering, we visualized the **normalized feature means** across clusters using a **radar chart**.

### Methodology

1. **Feature Aggregation**: We grouped the clustered dataset (`feature_data_clustered_v3.csv`) by `kmeans_cluster` and computed the mean value of each feature per cluster.
2. **Normalization**: For fair comparison, feature-wise min-max normalization was applied across clusters to scale all values to [0, 1].
3. **Radar Chart**: Each cluster is represented as a polygon over 30+ feature axes, highlighting relative strengths and weaknesses.

### Key Observations

- **Cluster 1** exhibits **higher values** in almost all features:
  - **Consumption metrics** (`avg_active_power`, `sum_sub_1/2/3`, etc.) are clearly elevated.
  - **Time-based variations** such as `lag_1~7`, `rolling_mean_7`, and `std_last_3_days` are also significantly higher.
  - Anomaly detection scores (`zscore`, `anomaly_zscore`, `anomaly_iforest`) are larger, suggesting more erratic or intense consumption behavior.
  - Even calendar-based features like `is_weekend` and `day_of_week` show slight variation.

- **Cluster 0**, in contrast, maintains relatively **low and stable** values across most features, indicating:
  - More regular and possibly lower baseline usage.
  - Fewer anomalies and more consistent load profiles.

### Interpretation

This analysis confirms that the clusters identified by unsupervised learning reflect **substantially different electricity usage behaviors**:

- **Cluster 1** likely includes households or periods with **high and fluctuating demand**, possibly aligned with seasonal peaks, weekends, or anomaly events.
- **Cluster 0** represents **more stable, low-consumption periods**, potentially corresponding to weekdays or off-peak seasons.

### Applications

These cluster profiles can be leveraged to:
- Tailor **load forecasting models per cluster**, improving overall accuracy.
- Design **demand response strategies** based on user type.
- Enhance **anomaly detection**, targeting only high-variability clusters.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load clustered dataset (v3)
df = pd.read_csv("/content/drive/MyDrive/electric_power_project/data/processed/feature_data_clustered_v3.csv")

# 2. Calculate per-cluster statistics
cluster_summary = df.groupby("kmeans_cluster").agg(["mean", "std"]).T

# 3. Normalize cluster means for radar chart visualization
cluster_means = df.groupby("kmeans_cluster").mean()
feature_cols = [col for col in cluster_means.columns if col != "target" and "cluster" not in col]
normalized_means = (cluster_means[feature_cols] - cluster_means[feature_cols].min()) / \
                   (cluster_means[feature_cols].max() - cluster_means[feature_cols].min())

# 4. Plot radar chart to visualize cluster consumption patterns
labels = feature_cols
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the circle

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

for i, row in normalized_means.iterrows():
    values = row.tolist()
    values += values[:1]
    ax.plot(angles, values, label=f'Cluster {i}')
    ax.fill(angles, values, alpha=0.1)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
plt.title("Cluster-wise Electricity Consumption Patterns (v3)", y=1.08)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()

# 5. Save plot
plt.savefig("/content/drive/MyDrive/electric_power_project/results/cluster_profiles_radar_v3.png")
plt.show()
```
